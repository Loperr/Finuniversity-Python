### Задание 2

На портале открытых данных Правительства Москвы получите и сохраните на диск набор данных <a href="https://data.mos.ru/opendata/624?pageSize=50&pageIndex=0&isDynamic=false"> Входы и выходы вестибюлей станций Московского метрополитена в csv формате</a> Копия этого набора данных нахдится в папке Data Вашего электронного учебного курса, имя файла *data-624-2999-01-01.csv*

Найдите все станции, являющиеся объектом культурного наследия. Распечатайте название станции, название линии и тип объекта культурного наследия (объект культурного наследия регионального значения или выявленный объект культурного наследия). Результат отсортируйте по имени станции.

````python
# 2.
import csv

with open('data-624-2999-01-01.csv', 'r', encoding = 'utf-8') as file:
    reader = csv.reader(file, delimiter=';')
    header = next(reader)
    data = list(reader)
    cultural_heritage_stations = []
    for row in data:
        if row[11] == 'объект культурного наследия регионального значения' or row[11] == 'выявленный объект культурного наследия':
            cultural_heritage_stations.append(row)
    for station in cultural_heritage_stations:
        cultural_heritage_stations.sort(key=lambda x: x[2])
        print(f"Название станции: {station[2]}, Название линии: {station[10]}, Тип объекта культурного наследия: {station[11]}")
````
